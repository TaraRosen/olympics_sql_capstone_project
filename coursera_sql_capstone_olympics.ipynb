```python
import pandas as pd

import sqlite3
```

```python
events_df = pd.read_csv('athlete_events.csv')
regions_df = pd.read_csv('noc_regions.csv')
```

```python
# connect to database
conn = sqlite3.connect(':memory:')

# create cursor
cur = conn.cursor()
```

```python
events_df.to_sql('events', con = conn, index=False)
regions_df.to_sql('regions', con = conn, index=False)
```

```python
cur.execute("""SELECT * 
               FROM events
            """)
events = pd.DataFrame(cur.fetchall())                    # create dataframe from sql query
events.columns = [x[0] for x in cur.description]         # labels dataframe columns
events                                                   # view dataframe
```

```python
print(events.Year.min())
print(events.Year.max())
```
